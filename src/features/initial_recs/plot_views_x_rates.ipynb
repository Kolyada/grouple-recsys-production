```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_ptn = '/data/groupLe_recsys/processed/{}/views.csv'
path_features_ptn = '/data/groupLe_recsys/interim/{}/items_features.csv'
```

```python
sites = 'dorama manga mint book'.split()
```

```python
# transforms
del_zeros = lambda df: df[~(df.rate == 0)]
```

# Explore dorama domain

```python
site = 'dorama'
```

```python
df = pd.read_csv(path_ptn.format(site))
df = del_zeros(df)
df.head(4)
```

```python
# build statistics
item2views = df.groupby('item_id')['rate'].count()
item2avg_rate = df.groupby('item_id')['rate'].sum()/df.groupby('item_id')['rate'].count()
```

```python
# plot views x avg rate
_ = plt.figure(figsize=(10, 5))
plt.scatter(item2views, item2avg_rate, s=1)
plt.xlabel('Views')
plt.ylabel('Avg rate')
# plt.xlim(-10, 2000)
```

```python
# plot views x avg rate
_ = plt.figure(figsize=(10, 5))
plt.scatter(item2views, item2avg_rate, s=1)
plt.xlabel('Views')
plt.ylabel('Avg rate')
plt.xlim(-10, 1200)
plt.ylim(5, 10.1)
```

```python
DORAMA_THRESHOLD = 1700
print('Total items', len(item2views))
print('Items above threshold', len(item2views[item2views>DORAMA_THRESHOLD]))
print('Percent of items above threshold', round(100*len(item2views[item2views>DORAMA_THRESHOLD])/len(item2views), 2), '%')
```

```python
# build mapping df
records = []
i2v = item2views.to_dict()
i2ar= item2avg_rate.to_dict()

for item_id, views in i2v.items():
    avg_rate = i2ar[item_id]
    records.append((item_id, views, avg_rate))

features_df = pd.DataFrame.from_records(records, columns='item_id views avg_rate'.split())
```

```python
features_df.to_csv(path_features_ptn.format(site), index=False)
```
